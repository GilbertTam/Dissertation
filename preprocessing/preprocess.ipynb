```python
#training set
import datetime

d_s = datetime.date(2020,3,3)   # train 2020-03-03 to 2021-06-14
d_e = datetime.date(2021,6,14)   # test 2021-06-15 2021-11-14

def getBetweenDay(d_s,d_e):
    date_list = []
    begin_date = datetime.datetime.strptime(d_s,"%Y-%m-%d")
    end_date = datetime.datetime.strptime(d_e,"%Y-%m-%d")
    while begin_date <= end_date:
        date_str = begin_date.strftime("%Y-%m-%d")
        date_list.append(date_str)
        begin_date += datetime.timedelta(days=1)
    return date_list


return_list = getBetweenDay('2020-03-03','2021-06-14')
#df = df_pred[(df_pred.loc[df_pred['Entity'] == "Israel"]) and (df_pred[df_pred['Day'].isin(return_list)])]
#df_pred = df_pred.loc[df_pred['Entity'] == "Israel"] 
master = master[master['Entity'] == "Israel"] 

master = master[master['Day'].isin(return_list)]
#master.reset_index
master.to_csv('./train/train_out.csv', index=False)
master
```
